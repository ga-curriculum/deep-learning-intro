In order to set up our neural network's architecture, PyTorch implementation requires you to build a class. This class will describe the layers of network and how data will propagate through the network.

Remember, a class is simply a template for an object or custom data type. This object will have specific attributes and behaviors.

Let's review some concepts real quick.

---

## Part 1. The Person Class

```python
class Person:

  # constructor method
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def print_name(self):
    print(f"Hi, my name is, {self.name}")

  def print_age(self):
    print(f"{self.name} is {self.age} years old.")
```

# Now You

Create an *instance* of the person class with your name and age, then invoke the `print_name` and `age` methods.

```python
# Answer:
new_person = Person("Shaq", 53)

new_person.print_name()

new_person.print_age()
```

# Inheritance
---
Inheritance allows one class to reuse and extend the methods and attributes of another class.
The original class is the parent class, and the new class that builds on it is the child class.

Why not put everything into one class? Inheritance helps us avoid duplication, keeps our code organized, and lets us modify or specialize behavior without changing the parent class. This makes our code more modular and easier to maintain.

In PyTorch, we’ll inherit from built-in PyTorch classes (like nn.Module) and extend them to define our own neural network architectures. Here’s a quick example of how inheritance works.

```python
# Let's make a student class that inherits from the person class
class Student(Person):
  def __init__(self, student_id, name, age):
    self.student_id = student_id
    # note the __super__, this tells python to pass the arg's from the parent's constructor
    super().__init__(name, age)

  def print_id(self):
    print(f"{self.name}'s id is {self.student_id}")
```

# Now You

Create an instance of student and invoke the `student_id` method and then a method from the `Person` class.

```python
# Answer:
stud_1 = Student(420, "Peter Parker", 16)

stud_1.print_id()

stud_1.print_age()
```
