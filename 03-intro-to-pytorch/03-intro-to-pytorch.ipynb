---

Tensors are very similar to arrays and matrices, we use them to encode the inputs and outputs of a model. For the most part, they will be similar to using `np.array`.

Why can't we use `np.array`?

Because Tensors are designed to run on hardware accelerators, like GPUs. When doing deep learning, it is likely that you'll have vast amounts of data, any techniques or data structures we can use to speed up our model training should be utilized.

# Datasets and DataLoaders
---

Data cleaning and processing is going to be the bulk of our work.

PyTorch provides two data primitives `torch.utils.data.DataLoader` and `torch.utils.data.Dataset`. We can use pre-loaded data sets and our own data.

What is the difference?

* `Dataset` stores the samples and their corresponding labels.
* `DataLoader` wraps an iterable around the `Dataset` to enable access to the samples.

We'll see an example with the [Fashion-MNIST](https://research.zalando.com/project/fashion_mnist/fashion_mnist/) data set.

---

Trying an example with a dummy data set.

```python
# First we need to set up a class that inherists from Dataset
class MyDataset(Dataset):
  def __init__(self, data, labels):
    self.data = data
    self.labels = labels

  def __len__(self):
    return len(self.data)

  def __getitem__(self, idx):
    x = self.data[idx]
    y = self.labels[idx]

    return x, y
```

---

Let's breakdown what's happening in our `MyDataset` class.

`Dataset` is an abstract base class. We need PyTorch to have two special methods, `__len__` and `__getitem__`.

The `__len__` method returns the size of the dataset (duh!). But why do we need that? The `DataLoader` uses this method to determine how many batches to make and when to start an epoch!

The `__getitem__` method fetches one sample, usually a feature and a label, from a given index (our `idx` argument). The `DataLoader` calls this to build batches for training.

The `DataLoader` uses our `DataSet` to make our model training more efficient!

---

So why can't we simply convert our dataset to a `tensor` and then throw it at `DataLoader`?

When we're working with *real* data out in the wild, we're likely to have lots of data than can't fit in memory. We'll need to preprocess and transform data on the fly. The `Dataset` gives you the power to define custom fetching logic and transformations.

## *Big Picture*

Big O notation and algorithmic complexity become important in deep learning. A good thing to keep in the back of your mind as a data scientist is how does your solution scale for more data. By wrapping our `Dataset` into the `DataLoader`, we're going to have an efficient and scalable pipeline. We're rarely going to have a dataset that will be a simple list or tensor that we'll pass through a model.
