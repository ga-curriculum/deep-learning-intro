# Exercise 1: Tensor Operations

1. Create a tensor with shape (4, 4) filled with random integers between 0 and 10.
2. Reshape it into a (2, 8) tensor.
3. Compute the mean of all elements in the reshaped tensor.
  **Hint** you may get an error if you don't specify the datatype in the method to calculate the mean

```python
import torch
```

```python
# Solution for Exercise 1
tensor = torch.randint(0, 10, (4, 4))
print("Original Tensor:")
print(tensor)

# Reshape
reshaped_tensor = tensor.view(2, 8)
print("Reshaped Tensor:")
print(reshaped_tensor)

# Compute mean
mean_value = reshaped_tensor.mean(dtype=float)
print("Mean of elements:", mean_value)
```

# Exercise 2:

Tensor Manipulation Create a tensor of shape (3, 3) with values from 1 to 9, perform slicing, and compute the sum of each row.

```python
tensor = torch.arange(1, 10).reshape(3, 3)
print("Tensor:")
print(tensor)

# Compute row-wise sum
row_sum = tensor.sum(dim=1)
print("Row-wise sum:", row_sum)
```

# Exercise 3: DataLoader Exploration

1. Create a dataset with 50 samples, where each sample has 3 features.
2. Use a DataLoader to batch the dataset into groups of 5 samples.
3. Print the first two batches.

```python
# imports
from torch.utils.data import DataLoader, TensorDataset
```

```python
# Solution for Exercise 2
features = torch.randn(50, 3)  # 50 samples, 3 features each
labels = torch.randint(0, 2, (50,))  # Binary labels

dataset = TensorDataset(features, labels)
data_loader = DataLoader(dataset, batch_size=5, shuffle=True)

# Print the first two batches
for i, (batch_features, batch_labels) in enumerate(data_loader):
    print(f"Batch {i + 1} Features:")
    print(batch_features)
    print(f"Batch {i + 1} Labels:")
    print(batch_labels)
    if i == 1:  # Stop after 2 batches
        break
```

# Exercise 4:

DataLoader Usage Create a DataLoader for a dataset with 100 samples and a batch size of 10. Shuffle the data and print the first batch.

```python
# Generate a larger dataset
features = torch.randn(100, 5)  # 100 samples, 5 features each
labels = torch.randint(0, 2, (100,))  # Binary labels

dataset = TensorDataset(features, labels)
data_loader = DataLoader(dataset, batch_size=10, shuffle=True)

# Print the first batch
for i, (batch_features, batch_labels) in enumerate(data_loader):
    if i == 0:
        print("First Batch Features:", batch_features)
        print("First Batch Labels:", batch_labels)
        break
```

Exercise 5: Custom Transform

1. Write a custom transform that doubles the value of each element in a tensor. **Hint**: This requires writing a custom class and will require a method called `__call__` instead of the usual constructor
2. Apply the transform to a dataset and verify the transformed values.

```python
# Solution for Exercise 3
class DoubleTransform:
    def __call__(self, tensor):
        return tensor * 2
```

```python
# Example tensor
example_tensor = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
transform = DoubleTransform()
```

```python
# Apply transform
transformed_tensor = transform(example_tensor)
print("Original Tensor:")
print(example_tensor)
print("Transformed Tensor:")
print(transformed_tensor)
```
