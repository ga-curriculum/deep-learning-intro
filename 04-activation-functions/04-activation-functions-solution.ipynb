### About
Provide a visualizatiobn of activation functions.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

Activation functions determine how data is transformed before being passed onto the next layer. They introduce non-linearity, which allows our neural network to learn complex patterns.

For the most part, our hidden layers will use ReLU. Some specific architectures, like RNNs, use different activation functions, but ReLU is the standard choice for DNN Hidden Layers.

Output layers are where activation functions matter the most. This will ensure our predictions will be mapped to the sample space of our target variable.

Quick reminder:

* Regression: Linear/Identity Activation (no transformation since output is continuous)
* Binary Classification: Sigmoid, maps numbers between 0-1
* Multi-Class Classification: Softmax, generalization of Sigmoid to more than one class, returns a vector of probabilities corresponding to each class.

```python
# Define x using torch linspace, 1000 equally spaced numbers in our interval
x = torch.linspace(-5, 5, 1000)

x[:-100]
```

```python
# let's use the activation functions and then plot them

relu = F.relu(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, relu)
plt.title("ReLU Activation");
```

```python
sigmoid = torch.sigmoid(x)

plt.plot(x, sigmoid)
plt.title("Sigmoid Activation");
```

```python
plt.plot(x, x)
plt.title("Linear/Identity Activation");
```

You'll sometimes see `tanh`, it's been used in some RNN's layers.

```python
tanh = torch.tanh(x)

plt.plot(x, tanh)
plt.title("tanh activation")
```
